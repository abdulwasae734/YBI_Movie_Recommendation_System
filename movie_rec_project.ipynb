MOVIE RECOMMENDATION SYSTEMS

Recommender System:is a system that is used to predict or filter preferences according to the user's choices.Recommender system is widely utilised in a variety of areas including movies,music,news,books,reasearch articles,search quieries,social tags and products in general. Recommended Systems produce a list of recommendations in in any of the two ways- Collabarative Filtering:Collabarative filtering approaches build a model from the user's past behaviour as well as similar decisions made by other users.This model is then used to predict the items that user's may have interest in. Content based Filtering:Content based filtering uses a series of discrete characteristics of an item in order to recommend items with similar properties

```python
x= df_features['Movie_Keywords']+' '+df_features['Movie_Genre']+' '+df_features['Movie_Language']+' '+df_features['Movie_Tagline']+' '+df_features['Movie_Director']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x = tfidf.fit_transform(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(x)
```

Get All Movies Sort Based on Recommendation Score wrt Favourite Movie
